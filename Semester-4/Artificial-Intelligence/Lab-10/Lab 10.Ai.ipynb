# Feature Engineering is the process of creating new input features or modifying existing ones to make them more useful for a machine learning model.
# in the titanic dataset Create a new feature by extracting titles like "Mr", "Mrs", etc. from the Name column.
# create a new feature family size by adding the sibsp and parch column values

```python

import re

def extract_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ""

titanic['Title'] = titanic['Name'].apply(extract_title)
```

create a new feature family size by adding the sibsp and parch column values

```python

titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
```
