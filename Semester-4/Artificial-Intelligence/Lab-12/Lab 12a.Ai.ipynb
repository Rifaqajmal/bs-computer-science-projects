#  **Minimax Algorithm in Rock-Paper-Scissors**

```python
import random
import time

class RPSGame:
    def __init__(self):
        self.moves = ['rock', 'paper', 'scissors']
        self.min_calls = 0
        self.max_calls = 0

    def get_winner(self, player_move, ai_move):
        if player_move == ai_move:
            return 0  # Tie
        elif (player_move == 'rock' and ai_move == 'scissors') or \
             (player_move == 'paper' and ai_move == 'rock') or \
             (player_move == 'scissors' and ai_move == 'paper'):
            return -1  # Player wins
        else:
            return 1  # AI wins

    def max(self):
        self.max_calls += 1
        max_score = -2
        best_move = None

        for ai_move in self.moves:
            (score, _) = self.min(ai_move)
            if score > max_score:
                max_score = score
                best_move = ai_move

        return max_score, best_move

    def min(self, ai_move):
        self.min_calls += 1
        min_score = 2
        best_move = None

        for player_move in self.moves:
            score = self.get_winner(player_move, ai_move)
            if score < min_score:
                min_score = score
                best_move = player_move

        return min_score, best_move

    def play(self):
        print("Welcome to Rock-Paper-Scissors!")
        print("Available moves: rock, paper, scissors")
        while True:
            player_move = input("Enter your move (or type 'quit' to exit): ").lower()
            if player_move == 'quit':
                break
            if player_move not in self.moves:
                print("Invalid move. Try again.")
                continue

            start = time.time()
            score, ai_move = self.max()
            end = time.time()

            print(f"AI chose: {ai_move}")
            result = self.get_winner(player_move, ai_move)
            if result == 0:
                print("It's a tie!")
            elif result == -1:
                print("You win!")
            else:
                print("AI wins!")

            print(f"Evaluation time: {round(end - start, 7)}s")
            print(f"Total max() calls: {self.max_calls}")
            print(f"Total min() calls: {self.min_calls}")
            print("-" * 30)

            # Reset counts for next round
            self.min_calls = 0
            self.max_calls = 0


def main():
    game = RPSGame()
    game.play()

if __name__ == "__main__":
    main()
```
