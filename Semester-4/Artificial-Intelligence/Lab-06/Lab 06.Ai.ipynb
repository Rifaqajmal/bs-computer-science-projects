#### what is a csv file format
#### A CSV file is a plain text file used to store data in a structured format.
#### It is one of the simplest and most common ways to store tabular data (like spreadsheets or tables).
#### The name "CSV" stands for Comma-Separated Values, which means the data is separated by commas.

### structure of csv file
#### Rows: Each line in the file represents a single row of data (like a row in a table).

#### columns: Each value in a row is separated by a comma, representing a column (like a field or attribute).

#### Header Row: The first row often contains the names of the columns (e.g., "Name, Age, City").

# Handling missing values/ Data Cleaning
## Detecting missing values
### isnull.sum

### fillna
### syntax = (value, method=None, axis=None, inplace=False, limit=None)
## Remove missing values
### dropna function (The dropna() method is used to remove rows or columns that contain missing values (NaN or None).)
### syntax (axis=0, how='any', thresh=None, subset=None, inplace=False)
### drop
### drop_duplicates
# handling categorical data


```python



# Detecting missing values
print(df.isnull().sum())

# Replacing missing values
# Example: Fill missing values in the 'Price' column with the mean
if 'Price' in df.columns:
    df['Price'].fillna(df['Price'].mean(), inplace=True)
# Example: Fill missing values in the 'RAM' column with the mode
if 'RAM' in df.columns:
    df['RAM'].fillna(df['RAM'].mode()[0], inplace=True)

# Removing rows with any missing values
df.dropna(inplace=True)

# Removing duplicate rows
df.drop_duplicates(inplace=True)

# Handling categorical data (example: one-hot encoding)
# Check if there are any categorical columns you want to encode
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0 :
    print("Categorical columns identified:")
    print(categorical_cols)
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print(df.head())
print(df.info())

```
