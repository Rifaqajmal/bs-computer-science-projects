```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    # Indented the return statement to be part of the function body
    return np.sin(5 * x) * (1 - np.tanh(x**2))

def hill_climb(f, x_start, max_iters=1000, step_size=0.01):
    x = x_start
    history = [x]
    for i in range(max_iters):
        left = x - step_size
        right = x + step_size

        if f(left) > f(x):
            x = left
        elif f(right) > f(x):
            x = right
        else:
            break
        history.append(x)
    return x, f(x), history

np.random.seed(12)
start = np.random.uniform(-2, 2)
print("starting point is :",start)
peak_x, peak_y, path = hill_climb(objective_function, start)
x_vals = np.linspace(-2, 2, 1000)
y_vals = objective_function(x_vals)
y_path = [objective_function(x) for x in path]
plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals, label="Objective Function", color='blue')
plt.plot(path, y_path, 'ro-', label='Search Path')
plt.scatter(peak_x, peak_y, color='green', s=100, label='Local Maximum')
plt.title("Hill Climbing Search Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Modify the algorithm to run from 10 different random starting points and return the
# best result.

import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return np.sin(5 * x) * (1 - np.tanh(x**2))

def hill_climb(f, x_start, max_iters=1000, step_size=0.01):
    x = x_start
    history = [x]
    for i in range(max_iters):
        left = x - step_size
        right = x + step_size

        if f(left) > f(x):
            x = left
        elif f(right) > f(x):
            x = right
        else:
            break
        history.append(x)
    return x, f(x), history

best_peak_x = None
best_peak_y = -float('inf') # Initialize with negative infinity

np.random.seed(12)
for _ in range(10):
    start = np.random.uniform(-2, 2)
    peak_x, peak_y, path = hill_climb(objective_function, start)

    if peak_y > best_peak_y:
        best_peak_x = peak_x
        best_peak_y = peak_y
        best_path = path

print("Best starting point is :",best_path[0])
print(f"Global Maximum found at x = {best_peak_x}, y = {best_peak_y}")

x_vals = np.linspace(-2, 2, 1000)
y_vals = objective_function(x_vals)
y_path = [objective_function(x) for x in best_path]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Objective Function", color='blue')
plt.plot(best_path, y_path, 'ro-', label='Search Path')
plt.scatter(best_peak_x, best_peak_y, color='green', s=100, label='Global Maximum')
plt.title("Hill Climbing Search Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

Task 4: Extend the algorithm to a 2D function, and visualize the surface + search path.

```python
# prompt: Task 4: Extend the algorithm to a 2D function, and visualize the surface + search path.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def objective_function_2d(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

def hill_climb_2d(f, x_start, y_start, max_iters=100, step_size=0.1):
    x, y = x_start, y_start
    history = [(x, y)]
    for _ in range(max_iters):
        best_x, best_y = x, y
        best_val = f(x,y)

        for dx in [-step_size, 0, step_size]:
            for dy in [-step_size, 0, step_size]:
                if dx == 0 and dy == 0:
                    continue
                nx, ny = x + dx, y + dy
                val = f(nx, ny)

                if val > best_val:
                  best_val = val
                  best_x, best_y = nx,ny

        if best_x == x and best_y == y:
            break
        x,y = best_x,best_y
        history.append((x, y))

    return x, y, f(x,y), history

# Generate the surface plot
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = objective_function_2d(X, Y)

# Perform hill climbing
start_x = np.random.uniform(-4, 4)
start_y = np.random.uniform(-4, 4)

peak_x, peak_y, peak_z, path = hill_climb_2d(objective_function_2d, start_x, start_y)

# Create the 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

# Plot the search path
path_x, path_y = zip(*path)
path_z = [objective_function_2d(px, py) for px, py in path]
ax.plot(path_x, path_y, path_z, marker='o', color='red', label='Search Path')

# Mark the local maximum
ax.scatter(peak_x, peak_y, peak_z, color='green', s=100, label='Local Maximum')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Hill Climbing in 2D')
ax.legend()
plt.show()
```

Task 5: Apply to a Real Problem: Maximize score in a simplified game or formula

```python
# prompt: Task 5: Apply to a Real Problem: Maximize score in a simplified game or formula

import numpy as np

def game_score(skill, effort):
    """
    A simplified game score function.
    Higher skill and effort generally lead to higher scores, but there's a diminishing returns effect.
    """
    return np.sqrt(skill) * np.log(1 + effort) - 0.1 * effort**2

def hill_climb_game(skill, max_iters=100, step_size=0.1):
    """
    Maximize game score by adjusting effort using hill climbing
    """
    effort = 0  # Start with zero effort
    history = [effort]
    best_effort = effort

    for i in range(max_iters):
        current_score = game_score(skill, effort)

        next_effort = effort + step_size
        next_score = game_score(skill, next_effort)
        if next_score > current_score:
            effort = next_effort
        else:
            # Try a smaller step in the other direction
            next_effort = effort - step_size
            next_score = game_score(skill, next_effort)
            if next_score > current_score:
                effort = next_effort

            else:
                break

        history.append(effort)
        if game_score(skill, effort) > game_score(skill, best_effort):
            best_effort = effort
    return best_effort, game_score(skill, best_effort), history

# Example usage:
skill_level = 5  # Example skill level
best_effort, max_score, effort_history = hill_climb_game(skill_level)

print(f"Skill Level: {skill_level}")
print(f"Best Effort: {best_effort}")
print(f"Maximum Score: {max_score}")

import matplotlib.pyplot as plt
plt.plot(effort_history, [game_score(skill_level, e) for e in effort_history])
plt.xlabel("Effort")
plt.ylabel("Score")
plt.title("Hill Climb to Maximize Game Score")
plt.show()
```
